```
# import libraries
import tensorflow as tf
import numpy as np 
```

```
# input & output
c_q = np.array([-40, -10,  0,  8, 15, 22,  38],  dtype=float)
f_a = np.array([-40,  14, 32, 46, 59, 72, 100],  dtype=float)
```

```
# model design , compile and fit

l0 = tf.keras.layers.Dense(units=1)

model = tf.keras.models.Sequential([l0])

model.compile(loss = 'mean_squared_error',optimizer = tf.keras.optimizers.Adam(0.1))

history = model.fit(c_q,f_a,epochs = 500)

```

```
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
```

```
l0.get_weights()
```

```
model.predict([100])
```

```
l0 = tf.keras.layers.Dense(units=4,input_shape = [1])
l1 = tf.keras.layers.Dense(units=4)
l2 = tf.keras.layers.Dense(units=1)



model = tf.keras.models.Sequential([l0,l1,l2])

model.compile(loss = 'mean_squared_error',optimizer = tf.keras.optimizers.Adam(0.1))

history = model.fit(c_q,f_a,epochs = 500)
```
